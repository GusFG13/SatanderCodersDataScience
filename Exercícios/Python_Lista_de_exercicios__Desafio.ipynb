ğŸ’¡ Entrada: 25.01 | SaÃ­da: (25,50]

1. Mora perto da vÃ­tima?
2. JÃ¡ trabalhou com a vÃ­tima?
3. Telefonou para a vÃ­tima?
4. Esteve no local do crime?
5. Devia para a vÃ­tima?

Cada resposta sim dÃ¡ um ponto para o suspeito. A polÃ­cia considera que os suspeitos com 5 pontos sÃ£o os assassinos, com 4 a 3 pontos sÃ£o cÃºmplices e 2 pontos sÃ£o apenas suspeitos, necessitando de outras investigaÃ§Ãµes. Valores abaixo de 2 sÃ£o liberados.

No seu programa, vocÃª deve fazer essas perguntas e, de acordo com as respostas do usuÃ¡rio, vocÃª vai informar como a polÃ­cia o considera.

```python
def resposta(pergunta):
    while True:
        print('Por favor, responda s para Sim ou n para NÃ£o')
        resp = input(pergunta)
        if resp in ['s', 'n']:
            if resp == 's':
                return 1
            else:
                return 0

perguntas = ['Mora perto da vÃ­tima?','JÃ¡ trabalhou com a vÃ­tima?','Telefonou para a vÃ­tima?','Esteve no local do crime?','Devia para a vÃ­tima?']
pontos = 0

for pergunta in perguntas:
    pontos = pontos + resposta(pergunta)
    
if pontos < 2:
    print(f'{pontos} ponto(s). Suspeito inocente.')
elif pontos == 2:
    print(f'{pontos} ponto(s). Suspeito ainda necessita investigaÃ§Ã£o.')
elif pontos == 3 or pontos == 4:
    print(f'{pontos} ponto(s). Suspeito provavelmente Ã© cÃºmplice.')
else:
    print(f'{pontos} ponto(s). Suspeito Ã© o assassino.')
```

Na mÃºsica, uma nota tem um tom (sua frequÃªncia, resultando em quÃ£o grave ou agudo Ã© o som) e uma duraÃ§Ã£o (por quanto tempo a nota soa). Neste problema, estamos interessados apenas na duraÃ§Ã£o das notas.

Marcos estÃ¡ dando os primeiros passos para ser um compositor de jingles. Ele estÃ¡ tendo alguns problemas, mas ao menos ele estÃ¡ criando melodias agradÃ¡veis e ritmos atrativos. Um jingle Ã© dividido em uma sequÃªncia de compassos, e um compasso Ã© formado de uma sÃ©rie de notas.

A duraÃ§Ã£o de uma nota Ã© indicada pela sua forma. Neste problema, iremos utilizar letras maiÃºsculas para indicar a duraÃ§Ã£o de uma nota. A seguinte tabela lista todas as notas disponÃ­veis:

A duraÃ§Ã£o de um compasso Ã© a soma da duraÃ§Ã£o de suas notas. Nos jingles de Marcos, cada compasso tem a mesma duraÃ§Ã£o. Como Marcos Ã© apenas um iniciante, seu famoso professor Johann Sebastian III o ensinou que a duraÃ§Ã£o de um compasso deve ser sempre 1.

Por exemplo, Marcos escreveu uma composiÃ§Ã£o contendo cinco compassos, dentre os quais quatro possuem a duraÃ§Ã£o correta e um estÃ¡ errado. No exemplo abaixo, cada compasso Ã© delimitado com barras e cada nota Ã© representada como na tabela acima.

Marcos gosta de computadores assim como de mÃºsica. Ele quer que vocÃª escreva um programa que determine, para cada uma de suas composiÃ§Ãµes, quantos compassos possuem a duraÃ§Ã£o correta e quais sÃ£o os compassos com duraÃ§Ã£o incorreta.

```python
#cria dicionÃ¡rio com a duraÃ§Ã£o de cada identificador
d = {'W':1,'H':1/2,'Q':1/4,'E':1/8,'S':1/16,'T':1/32,'X':1/64}

exemplo1 = '/HH/QQQQ/XXXTXTEQH/W/HW/'
exemplo2 = '/W/W/SQHES/'
exemplo3 = '/WE/TEX/THES/'

#alterar apenas aqui para trocar o exemplo
comp = exemplo3

incorretos = [] #lista de compassos com duraÃ§Ã£o incorreta

compassos = comp[1:(len(comp)-1)].split("/") #exclui as "/" inicial e final e separa os compassos em uma lista
#print(compassos)

n_compassos = len(compassos)

for compasso in compassos:
    duracao = 0
    for identificador in compasso: 
        duracao = duracao + d[identificador] #soma a duraÃ§Ã£o de cada identificador em um compasso
    #print(duracao)
    if duracao != 1:
        incorretos.append(compasso)

n_incorretos = len(incorretos)      
print(f'Qtd. de Corretos: {n_compassos - n_incorretos}')
if n_incorretos > 0:
    if n_incorretos == 1:
        print('Incorretos: ', incorretos[0])
    else:
        print('Incorretos: ', incorretos)
```

```python
from IPython.display import clear_output
from random import randrange
from time import sleep

def seleciona_palavra():
    arquivo = open('br-sem-acentos.txt', 'r',encoding="utf-8")
    linha = arquivo.read()
    arquivo.close()
    palavras = linha.split('\n')
    linha = ''
    while True:
        num_aleatorio = randrange(len(palavras))
        nova_palavra = palavras[num_aleatorio]
        if len(nova_palavra) >= 5: # palavras com no mÃ­nimo 5 letras
            palavras = []
            #print(f'{nova_palavra} : {len(nova_palavra)}')
            return nova_palavra
            #break

def escolheLetra():
    letra = input('Fale uma letra: ')
    while True:
        if len(letra) == 1:
            letra = letra.lower()
            if letra >= 'a' and letra <= 'z':
                return letra
            else:
                letra = input('Caractere invalido! Fale uma letra: ')
        else:
            print('Insiria apenda uma letra entre "a" e "z"')
            letra = input('Fale uma letra: ')

def imprime_forca(erros): 
    print(' +---+   ')
    print(' |   |   ')
    if erros > 0:
        print(' |   O   ')
        if erros > 1:
            if erros == 2:
                print(' |   |   ')
            elif erros == 3:
                print(' |  /|   ')
            else:
                print(' |  /|\  ')
                
            if erros > 4:
                if erros == 5:
                    print(' |  /    ')
                else:
                    print(' |  / \  ')
            else:
                print(' |       ')
        else:
            print(' |       ')
            print(' |       ')
    else:
        print(' |       ')
        print(' |       ')
        print(' |       ')
    print(' |       ')
    print('=========')
    
#InÃ­cio do jogo

nome = input('Entre seu nome: ')
clear_output()
print(f'OlÃ¡ {nome}! Seja bem vindo Ã  Forca!\nAdivinhe qual a palavra e fique livre.\nVocÃª deve falar uma letra por vez. Caso erre 6 vezes, serÃ¡ enforcado!!!')

sleep(5)
clear_output()

n_jogos = 0
n_vitorias = 0
maior_palavra = ''
compr_maior_palavra = 0


#Ã­nicio nova rodada
while True:
    letras_usadas = []
    lista_exibir = []
    n_erros = 0
    adivinhou = False
    n_jogos = n_jogos + 1

    while True:
        palavra = seleciona_palavra()
        #print(palavra)
        compr_palavra = len(palavra)
        for i in range(compr_palavra):
            lista_exibir.append('_')
        break
            
    print(f'A palavra chave tem {len(palavra)} letras')
    sleep(5)
    clear_output()

    while n_erros < 6 and not adivinhou:
        
        imprime_forca(n_erros)
        
        for i in range(len(lista_exibir)):
            print(f'{lista_exibir[i]} ', end='')
        print(f'\nLetras jÃ¡ usadas: {letras_usadas}')
        print(f'NÃºmero de tentativas erradas: {n_erros}')
        sleep(0.25)
        letra = escolheLetra()
        if letra in letras_usadas:
            print(f'Letra "{letra.upper()}" jÃ¡ usada. Escolha outra.')
            sleep(3)
            clear_output()
        else:
            letras_usadas.append(letra)
            if letra in palavra:
                for i in range(compr_palavra): #insere a letra nas posiÃ§Ãµes em que ela existe na palavra
                    if palavra[i] == letra:
                        lista_exibir[i] = letra
                if '_' not in lista_exibir: #encontrou todas as letras
                    adivinhou = True
            else: #letra errada
                n_erros = n_erros + 1
            clear_output()
    if adivinhou:
        n_vitorias = n_vitorias + 1
        if compr_palavra >= compr_maior_palavra:
            compr_maior_palavra = compr_palavra
            maior_palavra = palavra
        print(f'ParabÃ©ns {nome}!!! VocÃª descobriu a palavra {palavra.upper()}.')  
    else:
        imprime_forca(n_erros)
        print(f'Ah nÃ£o, {nome}!!! VocÃª foi enforcado. A palavra era {palavra.upper()}.')
    sleep(5)
    clear_output()
    #Nova rodada ou encerra jogo
    if 's' != input('Deseja jogar novamente? (s para sim, qualquer outra letra para encerrar)'):
        break
    else:
        clear_output()
        sleep(0.25)
#Fim de jogo e estatÃ­sticas
clear_output()
print(f'{nome}, vocÃª jogou {n_jogos} vezes e acertou {n_vitorias} palavra(s) ({((n_vitorias * 100) / n_jogos):.2f}% de acertos).')
if n_vitorias > 0:
    print(f'A maior palavra que vocÃª acertou foi {maior_palavra.upper()} com {compr_maior_palavra} letras.')
print('Obrigado por jogar!!')
```
