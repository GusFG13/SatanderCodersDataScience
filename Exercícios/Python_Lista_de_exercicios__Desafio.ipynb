Questão 2.
Crie um programa que leia um valor qualquer e apresente uma mensagem dizendo em qual dos seguintes intervalos ([0,25], (25,50], (50,75], (75,100]) este valor se encontra. Caso o valor não esteja em nenhum destes intervalos, deverá ser impressa a mensagem “Fora de intervalo”. Veja alguns exemplo abaixo:

💡 Entrada: 25.01 | Saída: (25,50]
Entrada: 25.00 | Saída: [0,25]
Entrada: 100.00 | Saída: (75,100]
Entrada: -25.02 | Saída: Fora de intervalo

📌Lembrando que o [ ou ] representa que o valor está contido no intervalo, enquanto o ( ou ) representa que o valor associado não está contido no intervalo. Em outras palavras, (75, 100] representa o intervalo que vai de maior que 75 (não incluindo o 75) até menor ou igual 100.

```python
numero = input('Informe um número: ')
numero = float(numero)
if numero >= 0 and numero <= 25:
    print('[0,25]')
elif numero > 25 and numero <= 50:
    print('(25,50]')
elif numero > 50 and numero <= 75:
    print('(50,75]')
elif numero > 75 and numero <= 100:
    print('(75,100]')
else:
    print('Fora de intervalo')
```

a) o produto entre o dobro do primeiro e a metade do segundo.

Questão 5.
Vamos fazer um programa para verificar quem é o assassino de um crime. Para descobrir o assassino, a polícia faz um pequeno questionário com 5 perguntas onde a resposta só pode ser sim ou não:

1. Mora perto da vítima?
2. Já trabalhou com a vítima?
3. Telefonou para a vítima?
4. Esteve no local do crime?
5. Devia para a vítima?

Cada resposta sim dá um ponto para o suspeito. A polícia considera que os suspeitos com 5 pontos são os assassinos, com 4 a 3 pontos são cúmplices e 2 pontos são apenas suspeitos, necessitando de outras investigações. Valores abaixo de 2 são liberados.

No seu programa, você deve fazer essas perguntas e, de acordo com as respostas do usuário, você vai informar como a polícia o considera.

```python
def resposta(pergunta):
    while True:
        print('Por favor, responda s para Sim ou n para Não')
        resp = input(pergunta)
        if resp in ['s', 'n']:
            if resp == 's':
                return 1
            else:
                return 0

perguntas = ['Mora perto da vítima?','Já trabalhou com a vítima?','Telefonou para a vítima?','Esteve no local do crime?','Devia para a vítima?']
pontos = 0

for pergunta in perguntas:
    pontos = pontos + resposta(pergunta)
    
if pontos < 2:
    print(f'{pontos} ponto(s). Suspeito inocente.')
elif pontos == 2:
    print(f'{pontos} ponto(s). Suspeito ainda necessita investigação.')
elif pontos == 3 or pontos == 4:
    print(f'{pontos} ponto(s). Suspeito provavelmente é cúmplice.')
else:
    print(f'{pontos} ponto(s). Suspeito é o assassino.')
```

Questão 10.
Na música, uma nota tem um tom (sua frequência, resultando em quão grave ou agudo é o som) e uma duração (por quanto tempo a nota soa). Neste problema, estamos interessados apenas na duração das notas.

Marcos está dando os primeiros passos para ser um compositor de jingles. Ele está tendo alguns problemas, mas ao menos ele está criando melodias agradáveis e ritmos atrativos. Um jingle é dividido em uma sequência de compassos, e um compasso é formado de uma série de notas.

A duração de uma nota é indicada pela sua forma. Neste problema, iremos utilizar letras maiúsculas para indicar a duração de uma nota. A seguinte tabela lista todas as notas disponíveis:

A duração de um compasso é a soma da duração de suas notas. Nos jingles de Marcos, cada compasso tem a mesma duração. Como Marcos é apenas um iniciante, seu famoso professor Johann Sebastian III o ensinou que a duração de um compasso deve ser sempre 1.

Por exemplo, Marcos escreveu uma composição contendo cinco compassos, dentre os quais quatro possuem a duração correta e um está errado. No exemplo abaixo, cada compasso é delimitado com barras e cada nota é representada como na tabela acima.

Marcos gosta de computadores assim como de música. Ele quer que você escreva um programa que determine, para cada uma de suas composições, quantos compassos possuem a duração correta e quais são os compassos com duração incorreta.

```python
#cria dicionário com a duração de cada identificador
d = {'W':1,'H':1/2,'Q':1/4,'E':1/8,'S':1/16,'T':1/32,'X':1/64}

exemplo1 = '/HH/QQQQ/XXXTXTEQH/W/HW/'
exemplo2 = '/W/W/SQHES/'
exemplo3 = '/WE/TEX/THES/'

#alterar apenas aqui para trocar o exemplo
comp = exemplo3

incorretos = [] #lista de compassos com duração incorreta

compassos = comp[1:(len(comp)-1)].split("/") #exclui as "/" inicial e final e separa os compassos em uma lista
#print(compassos)

n_compassos = len(compassos)

for compasso in compassos:
    duracao = 0
    for identificador in compasso: 
        duracao = duracao + d[identificador] #soma a duração de cada identificador em um compasso
    #print(duracao)
    if duracao != 1:
        incorretos.append(compasso)

n_incorretos = len(incorretos)      
print(f'Qtd. de Corretos: {n_compassos - n_incorretos}')
if n_incorretos > 0:
    if n_incorretos == 1:
        print('Incorretos: ', incorretos[0])
    else:
        print('Incorretos: ', incorretos)
```

Lista de palavras sem acento disponível em https://www.ime.usp.br/~pf/dicios/br-sem-acentos.txt

```python
from IPython.display import clear_output
from random import randrange
from time import sleep

def seleciona_palavra():
    arquivo = open('br-sem-acentos.txt', 'r',encoding="utf-8")
    linha = arquivo.read()
    arquivo.close()
    palavras = linha.split('\n')
    linha = ''
    while True:
        num_aleatorio = randrange(len(palavras))
        nova_palavra = palavras[num_aleatorio]
        if len(nova_palavra) >= 5: # palavras com no mínimo 5 letras
            palavras = []
            #print(f'{nova_palavra} : {len(nova_palavra)}')
            return nova_palavra
            #break

def escolheLetra():
    letra = input('Fale uma letra: ')
    while True:
        if len(letra) == 1:
            letra = letra.lower()
            if letra >= 'a' and letra <= 'z':
                return letra
            else:
                letra = input('Caractere invalido! Fale uma letra: ')
        else:
            print('Insiria apenda uma letra entre "a" e "z"')
            letra = input('Fale uma letra: ')

def imprime_forca(erros): 
    print(' +---+   ')
    print(' |   |   ')
    if erros > 0:
        print(' |   O   ')
        if erros > 1:
            if erros == 2:
                print(' |   |   ')
            elif erros == 3:
                print(' |  /|   ')
            else:
                print(' |  /|\  ')
                
            if erros > 4:
                if erros == 5:
                    print(' |  /    ')
                else:
                    print(' |  / \  ')
            else:
                print(' |       ')
        else:
            print(' |       ')
            print(' |       ')
    else:
        print(' |       ')
        print(' |       ')
        print(' |       ')
    print(' |       ')
    print('=========')
    
#Início do jogo

nome = input('Entre seu nome: ')
clear_output()
print(f'Olá {nome}! Seja bem vindo à Forca!\nAdivinhe qual a palavra e fique livre.\nVocê deve falar uma letra por vez. Caso erre 6 vezes, será enforcado!!!')

sleep(5)
clear_output()

n_jogos = 0
n_vitorias = 0
maior_palavra = ''
compr_maior_palavra = 0


#ínicio nova rodada
while True:
    letras_usadas = []
    lista_exibir = []
    n_erros = 0
    adivinhou = False
    n_jogos = n_jogos + 1

    while True:
        palavra = seleciona_palavra()
        #print(palavra)
        compr_palavra = len(palavra)
        for i in range(compr_palavra):
            lista_exibir.append('_')
        break
            
    print(f'A palavra chave tem {len(palavra)} letras')
    sleep(5)
    clear_output()

    while n_erros < 6 and not adivinhou:
        
        imprime_forca(n_erros)
        
        for i in range(len(lista_exibir)):
            print(f'{lista_exibir[i]} ', end='')
        print(f'\nLetras já usadas: {letras_usadas}')
        print(f'Número de tentativas erradas: {n_erros}')
        sleep(0.25)
        letra = escolheLetra()
        if letra in letras_usadas:
            print(f'Letra "{letra.upper()}" já usada. Escolha outra.')
            sleep(3)
            clear_output()
        else:
            letras_usadas.append(letra)
            if letra in palavra:
                for i in range(compr_palavra): #insere a letra nas posições em que ela existe na palavra
                    if palavra[i] == letra:
                        lista_exibir[i] = letra
                if '_' not in lista_exibir: #encontrou todas as letras
                    adivinhou = True
            else: #letra errada
                n_erros = n_erros + 1
            clear_output()
    if adivinhou:
        n_vitorias = n_vitorias + 1
        if compr_palavra >= compr_maior_palavra:
            compr_maior_palavra = compr_palavra
            maior_palavra = palavra
        print(f'Parabéns {nome}!!! Você descobriu a palavra {palavra.upper()}.')  
    else:
        imprime_forca(n_erros)
        print(f'Ah não, {nome}!!! Você foi enforcado. A palavra era {palavra.upper()}.')
    sleep(5)
    clear_output()
    #Nova rodada ou encerra jogo
    if 's' != input('Deseja jogar novamente? (s para sim, qualquer outra letra para encerrar)'):
        break
    else:
        clear_output()
        sleep(0.25)
#Fim de jogo e estatísticas
clear_output()
print(f'{nome}, você jogou {n_jogos} vezes e acertou {n_vitorias} palavra(s) ({((n_vitorias * 100) / n_jogos):.2f}% de acertos).')
if n_vitorias > 0:
    print(f'A maior palavra que você acertou foi {maior_palavra.upper()} com {compr_maior_palavra} letras.')
print('Obrigado por jogar!!')
```
